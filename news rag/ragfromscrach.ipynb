```python
from news_loader import fetch_news
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain_community.vectorstores import FAISS
import os

os.makedirs("vectorstores",exist_ok=True)

news_docs=fetch_news(
    query="stock market or inflation or interest rates or earnings"
)

print("News articles:",len(news_docs))

splitter=RecursiveCharacterTextSplitter(chunk_size=400,chunk_overlap=50)

chunks=[]

for doc in news_docs:
    chunks.extend(splitter.split_text(doc))

print("text chunks:",len(chunks))

embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

news_db=FAISS.from_texts(chunks,embeddings)
news_db.save_local("vectorstores/news_store")
print("news vectorstore created")
```
